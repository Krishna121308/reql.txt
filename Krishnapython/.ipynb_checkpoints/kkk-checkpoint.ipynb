```python
import numpy as np
#FUNCTIONS USED TO INITIALIZE ARRAYS IN NUMPY
array=np.ones((3,7)) # CREATES ARRAY WITH ELEMENTS (1)
print(array)
array_2=np.random.random([3,3]) # CREATES ARRAY WITH ELEMENTS (0 TO 1)
print(array_2)
array_3=np.arange(1,102,10) # CREATES ARRAY WITH ELEMENTS (FROM 1 TO 100 (101 EXCLUDEDprint(array_3)
array_4=np.linspace(10,16,25) # CREATES ARRAY WITH ELEMENTS (FROM 1 TO 14 25 ELEMENTS print(array_4)
array_5=np.full((3,3),20)# CREATES ARRAY WITH ELEMENTS (20 ONLY)
print(array_5)
array_6=np.array([1,2,3]) # REPEATS ARRAY 2 TIMES AND CREATES A 1D ARRAY
print(np.tile(array_6,2))
print(np.eye(3)) # CREATES IDENTITY MATRIX WITH SIZE = 3
print(np.random.randint(1,11,(3,3))) 
```

```python
array_1d = np.arange(1,21)
# fifth element
print(array_1d[5])
# Specific elements
# Notice that array[4, 7, 8] will throw an error, you need to provide the indices as aprint(array_1d[[2, 5, 6]])
# Slice third element onwards
print(array_1d[5:])
# Slice first three elements
print(array_1d[:9])
# Slice third to seventh elements
print(array_1d[1:11])
# Subset starting 0 at increment of 2

print(array_1d[0::2])
```

```python
# Creating a 2-D array
array_2d = np.array([[2, 5, 7, 5], [5,7,9,2], [18,15,16,19]])
# Third row second column 
print(array_2d[2, 1])
# Slicing the second row, and all columns
# Notice that the resultant is itself a 1-D array
print(array_2d[1, :])
# Slicing all rows and the third column 
print(array_2d[:, 2])
# Slicing all rows and the first three columns
print(array_2d[:, :3])
```

```python
## Comparing time taken for computation 
list_1 = [i for i in range(1000000)]
list_2 = [j**2 for j in range(1000000)]

# list multiplication 
import time

# store start time, time after computation, and take the difference 
t0 = time.time()
product_list = list(map(lambda x, y: x*y, list_1, list_2))
t1 = time.time()
list_time = t1 - t0 
print(t1-t0)
# numpy array
array_1 = np.array(list_1) 
array_2 = np.array(list_2)
t0 = time.time()
array_3 = array_1*array_2 
t1 = time.time()
numpy_time = t1 - t0 
print(t1-t0)
print("The ratio of time taken is {}".format(list_time/numpy_time))
```

```python
import numpy as np
# Reshape a 1-D array to a 3 x 4 array 
some_array = np.arange(0, 12).reshape(3, 4) 
print(some_array)


#creating two arrays

array_1 = np.arange(12).reshape(3, 4) 
array_2 = np.arange(20).reshape(5, 4)
print(array_1)
print("\n")
print(array_2) # vstack
# Note that np.vstack(a, b) throws an error - you need to pass the arrays as a list
np.vstack((array_1, array_2))

```

```python
# Basic mathematical operations 
a = np.arange(1, 20)
# sin, cos, exp, log 
print(np.sin(a))
print(np.cos(a))
print(np.exp(a))
print(np.log(a))
```

```python
# # Basic mathematical operations
a = np.arange(1, 20)
print(a)
# The non-numpy way, not recommended
a_list = [x/(x+1) for x in a]
print(a_list)
# The numpy way: vectorize the function, then apply it
f = np.vectorize(lambda x: x/(x+1))
print(f(a))
# Apply function on a 2-d array: Applied to each element
b = np.linspace(1, 100, 10)
print(f(b))
```

```python
# Creating arrays
a = np.arange(1, 10).reshape(3, 3)
b= np.arange(1, 13).reshape(3, 4) 
print(a)
print(b) # Inverse
print(np.linalg.inv(a)) # Determinant
print(np.linalg.det(a))
# Eigenvalues and eigenvectors 
print(np.linalg.eig(a))
# Multiply matrices
print(np.dot(a, b))
```

```python
import pandas as pd

# DataFrames with same columns
df1 = pd.DataFrame({
    'Name': ['Aman', 'Joy', 'Rashmi', 'Saif'],
    'Age': ['34', '31', '22', '33'],
    'Gender': ['M', 'M', 'F', 'M']
})

df2 = pd.DataFrame({
    'Name': ['Akhil', 'Asha', 'Preeti'],
    'Age': ['31', '22', '23'],
    'Gender': ['M', 'F', 'F']
})

# Concatenate row-wise
combined_rows = pd.concat([df1, df2], axis=0)
print(combined_rows)
```

```python
import pandas as pd

# IPL 2018 stats
IPL_2018 = pd.DataFrame({
    'IPL Team': ['CSK', 'SRH', 'KKR', 'RR', 'MI', 'RCB', 'KXIP'],
    'Matches Played': [16, 17, 16, 15, 14, 14, 14],
    'Matches Won': [11, 10, 9, 7, 6, 6, 6]
})
IPL_2018.set_index('IPL Team', inplace=True)

# IPL 2017 stats
IPL_2017 = pd.DataFrame({
    'IPL Team': ['MI', 'RPS', 'KKR', 'SRH', 'KXIP', 'DD'],
    'Matches Played': [17, 16, 16, 15, 14, 14],
    'Matches Won': [12, 10, 9, 8, 7, 6]
})
IPL_2017.set_index('IPL Team', inplace=True)

# Combine stats using addition
Total = IPL_2018.add(IPL_2017, fill_value=0)
print(Total)
```
